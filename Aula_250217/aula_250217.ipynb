```python
def process_sdg_data(csv_file, columns_to_drop):
    df = pd.read_csv(csv_file)
    df = df.drop(columns_to_drop, axis=1)
    df = df.set_index("Country Name").transpose()
    return df  
```

```python
df = process_sdg_data("arquivos\\sdg_literacy_rate.csv", ["Series Name", "Series Code", "Country Code"])
df2 = process_sdg_data("arquivos\\sdg_electricity_data.csv", ["Series Name", "Series Code", "Country Code"])
df3 = process_sdg_data("arquivos\\sdg_urban_population.csv", ["Series Name", "Series Code", "Country Code"])
```
